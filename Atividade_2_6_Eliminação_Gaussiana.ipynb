```python
import numpy as np
from scipy.linalg import solve

def gaussian_elimination_partial_pivot(A, b, tol=1e-12, verbose=False):
    """
    Eliminação gaussiana com pivotamento parcial.
    Retorna o vetor solução x.
    Se verbose=True imprime os passos intermediários.
    """
    A = A.astype(float).copy()
    b = b.astype(float).copy()
    n = len(b)

    for k in range(n-1):
        # Escolhe linha do maior pivô absoluto
        max_row = np.argmax(np.abs(A[k:, k])) + k
        if abs(A[max_row, k]) < tol:
            raise ValueError("Matriz singular ou quase singular.")

        # Troca de linhas se necessário
        if max_row != k:
            A[[k, max_row]] = A[[max_row, k]]
            b[[k, max_row]] = b[[max_row, k]]
            if verbose:
                print(f"\nTroca de linhas {k} <-> {max_row}")
                print("A:\n", A)
                print("b:", b)

        # Eliminação
        pivot = A[k, k]
        for i in range(k+1, n):
            m = A[i, k] / pivot
            A[i, k:] -= m * A[k, k:]
            b[i] -= m * b[k]
            if verbose:
                print(f"\nEliminando linha {i}, m={m}")
                print("A:\n", A)
                print("b:", b)

    # Substituição retroativa
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        if abs(A[i, i]) < tol:
            raise ValueError("Zero na diagonal durante substituição retroativa.")
        x[i] = (b[i] - np.dot(A[i, i+1:], x[i+1:])) / A[i, i]

    if verbose:
        print("\nMatriz escalonada (U):\n", A)
        print("Vetor b modificado:", b)
        print("Solução encontrada:", x)

    return x


# ========================
# Teste no sistema (a)
# ========================

A = np.array([[1, 1, 1],
              [4, 4, 2],
              [2, 1,-1]], dtype=float)

b = np.array([1, 2, 0], dtype=float)

# Solução com nosso método
x_gauss = gaussian_elimination_partial_pivot(A, b, verbose=True)

# Solução com scipy
x_scipy = solve(A, b)

print("\n==============================")
print("Solução pelo nosso método:", x_gauss)
print("Solução pelo scipy.linalg.solve:", x_scipy)
print("==============================")
```
