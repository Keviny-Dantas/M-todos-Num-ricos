ATIVIDADE 3.1 - Decomposição LU


Import das bibliotecas que serão utilizadas


```python
import numpy as np
from scipy.linalg import lu, solve_triangular
import timeit

```

Implementação das funções LU e Gaussiana

```python
def solve_lu_decomposition(A, b):
    P, L, U = lu(A)
    Pb = P @ b
    y = solve_triangular(L, Pb, lower=True)
    x = solve_triangular(U, y, lower=False)
    return x
```

```python
def solve_gaussian_elimination(A_in, b_in):
    A = np.copy(A_in).astype(float)
    b = np.copy(b_in).astype(float)
    n = len(b)

    for i in range(n):
        max_row = i
        for k in range(i + 1, n):
            if abs(A[k, i]) > abs(A[max_row, i]):
                max_row = k

        A[[i, max_row]] = A[[max_row, i]]
        b[[i, max_row]] = b[[max_row, i]]

        if A[i, i] == 0:
            raise ValueError("Matriz singular detectada")

        for k in range(i + 1, n):
            factor = A[k, i] / A[i, i]
            A[k, i:] -= factor * A[i, i:]
            b[k] -= factor * b[i]

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - np.dot(A[i, i + 1:], x[i + 1:])) / A[i, i]

    return x
```

Testes de Decomposições com LU

```python
A_a = np.array([[2.0, 1.0], [1.0, 2.0]])
b_a = np.array([1.0, 2.0])
sol_a_esperada = np.array([0.0, 1.0])

x_a = solve_lu_decomposition(A_a, b_a)
print(f"Teste (a): {A_a.tolist()} x = {b_a.tolist()}")
print(f"Solução encontrada: {x_a}")
print(f"Solução esperada:   {sol_a_esperada}")
print(f"Correto: {np.allclose(x_a, sol_a_esperada)}\n")

A_b = np.array([[4.0, 1.0], [1.0, 5.0]])
b_b = np.array([2.0, 10.0])
sol_b_esperada = np.array([0.0, 2.0])

x_b = solve_lu_decomposition(A_b, b_b)
print(f"Teste (b): {A_b.tolist()} x = {b_b.tolist()}")
print(f"Solução encontrada: {x_b}")
print(f"Solução esperada:   {sol_b_esperada}")
print(f"Correto: {np.allclose(x_b, sol_b_esperada)}\n")
```
