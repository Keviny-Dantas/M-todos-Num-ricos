Exercício 2-1 (Soma e Subtração de Vetores)

```python
print("--- Exercício 2-1: Plotagem de Vetores ---")

v = np.array([1, 2])
w = np.array([4, -3])

v_plus_w = v + w
v_minus_w = v - w

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico A: Soma
ax1.set_title('A) Vetores v, w, e v+w')
ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='k', label='v')
ax1.quiver(v[0], v[1], w[0], w[1], angles='xy', scale_units='xy', scale=1, color='dimgray', label='w')
ax1.quiver(0, 0, v_plus_w[0], v_plus_w[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='v+w')
ax1.legend()

# Gráfico B: Subtração
ax2.set_title('B) Vetores v, w, e v-w')
ax2.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='k', label='v')
ax2.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='dimgray', label='w')
ax2.quiver(w[0], w[1], v_minus_w[0], v_minus_w[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='v-w')
ax2.legend()

# Configurações dos eixos
for ax in [ax1, ax2]:
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.grid(linestyle='--')
    ax.set_aspect('equal', adjustable='box')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

plt.show()
```

Exercício 2-2 (Cálculo da Norma)

```python
print("\n--- Exercício 2-2: Algoritmo da Norma Euclidiana ---")

def calcular_norma(vetor):
    soma_quadrados = 0
    # O flatten garante que funcione para vetores linha e coluna
    for elemento in vetor.flatten():
        soma_quadrados += elemento**2
    return np.sqrt(soma_quadrados)

# Testes
v1 = np.array([1, 2, 3, 4, 5])
print(f"Vetor 1D: {v1}")
print(f"Minha função: {calcular_norma(v1):.5f}")
print(f"np.linalg.norm(): {np.linalg.norm(v1):.5f}\n")

v2 = np.array([[1, 2, 3]])
print(f"Vetor Linha: {v2} (shape: {v2.shape})")
print(f"Minha função: {calcular_norma(v2):.5f}")
print(f"np.linalg.norm(): {np.linalg.norm(v2):.5f}\n")

v3 = np.array([[4], [5], [6]])
print(f"Vetor Coluna: \n{v3} (shape: {v3.shape})")
print(f"Minha função: {calcular_norma(v3):.5f}")
print(f"np.linalg.norm(): {np.linalg.norm(v3):.5f}\n")

v_rand = np.random.randn(1, 10)
print(f"Vetor Aleatório (shape: {v_rand.shape})")
print(f"Minha função: {calcular_norma(v_rand):.5f}")
print(f"np.linalg.norm(): {np.linalg.norm(v_rand):.5f}\n")
```

Exercícios 2-3 e 2-4 (Vetores Unitários e Escalonamento)

```python
print("--- Exercício 2-3: Vetor Unitário ---")

def criar_vetor_unitario(vetor):
    norma = np.linalg.norm(vetor)
    if norma == 0:
        print("Aviso: O vetor nulo foi inserido. Retornando vetor original.")
        return vetor
    return vetor / norma

v = np.array([2, 4, 1])
unit_v = criar_vetor_unitario(v)
print(f"Vetor Original: {v}")
print(f"Vetor Unitário: {unit_v}")
print(f"Norma do Unitário: {np.linalg.norm(unit_v):.4f}\n")

# Teste com vetor nulo
zeros_v = np.array([0, 0, 0])
print(f"Teste Nulo: {criar_vetor_unitario(zeros_v)}\n")


print("--- Exercício 2-4: Escalar Vetor ---")

def escalar_vetor(vetor, nova_magnitude):
    norma_original = np.linalg.norm(vetor)
    if norma_original == 0:
        return vetor
    vetor_unitario = vetor / norma_original
    return vetor_unitario * nova_magnitude

v = np.array([3, 1, 5])
mag_desejada = 10
v_novo = escalar_vetor(v, mag_desejada)

print(f"Novo Vetor (Mag {mag_desejada}): {v_novo}")
print(f"Norma do Novo Vetor: {np.linalg.norm(v_novo):.4f}")
```

Exercício 2-5 (Transposição com Loop)

```python
print("\n--- Exercício 2-5: Transposição Manual ---")

vetor_linha = np.array([[10, 20, 30, 40, 50]])
num_elementos = vetor_linha.shape[1]
vetor_coluna = np.zeros((num_elementos, 1))

for i in range(num_elementos):
    vetor_coluna[i, 0] = vetor_linha[0, i]

print("Vetor Linha Original:")
print(vetor_linha)
print("Vetor Coluna Transposto:")
print(vetor_coluna)
```

Exercícios 2-6 e 2-7 (Produto Escalar)

```python
print("\n--- Exercício 2-6: Norma ao Quadrado vs Dot Product ---")
v = np.random.randn(5)
norma_sq = np.linalg.norm(v)**2
dot_prod = np.dot(v, v)

print(f"Vetor: {v}")
print(f"Norma^2: {norma_sq:.6f}")
print(f"Dot(v,v): {dot_prod:.6f}")
print(f"Iguais? {np.isclose(norma_sq, dot_prod)}\n")

print("--- Exercício 2-7: Comutatividade ---")
a = np.random.randn(10)
b = np.random.randn(10)
dot_ab = np.dot(a, b)
dot_ba = np.dot(b, a)
print(f"Dot(a,b): {dot_ab:.6f}")
print(f"Dot(b,a): {dot_ba:.6f}")
print(f"Iguais? {np.isclose(dot_ab, dot_ba)}")
```

Exercício 2-8 (Projeção Vetorial)

```python
print("\n--- Exercício 2-8: Visualização da Projeção ---")

a = np.array([2, 5])
b = np.array([4, 1])

# Cálculo da projeção
beta = np.dot(a, b) / np.dot(a, a)
proj_b_em_a = beta * a
vetor_distancia = b - proj_b_em_a

plt.figure(figsize=(7, 7))
plt.arrow(0, 0, a[0], a[1], head_width=.2, width=.05, color='k', length_includes_head=True, label='a')
plt.arrow(0, 0, b[0], b[1], head_width=.2, width=.05, color='k', length_includes_head=True, label='b')

# Plot da projeção e linha tracejada
plt.plot([0, proj_b_em_a[0]], [0, proj_b_em_a[1]], 'k', marker='o', label='beta*a')
plt.plot([proj_b_em_a[0], b[0]], [proj_b_em_a[1], b[1]], 'k--', label='dist')

# Texto
plt.text(a[0]+0.2, a[1], 'a', fontsize=12, fontweight='bold')
plt.text(b[0]+0.2, b[1], 'b', fontsize=12, fontweight='bold')

plt.grid(linestyle='--')
plt.axis('equal') # Importante para ver a ortogonalidade
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.legend()
plt.title("Projeção de b em a")
plt.show()
```

Exercícios 2-9 e 2-10 (Decomposição Ortogonal e Debugging)

```python
print("--- Exercício 2-9: Decomposição Ortogonal ---")

t = np.random.randn(2)
r = np.random.randn(2)

# Decomposição
t_paralelo = r * (np.dot(t, r) / np.dot(r, r))
t_perp = t - t_paralelo

print(f"Soma correta? {np.allclose(t, t_paralelo + t_perp)}")
print(f"Ortogonal? (Dot deve ser 0): {np.dot(t_perp, r):.10f}")

# Plotagem
plt.figure(figsize=(6, 6))
plt.arrow(0, 0, t[0], t[1], head_width=.05, color='k', length_includes_head=True, label='t')
plt.arrow(0, 0, r[0], r[1], head_width=.05, color='gray', length_includes_head=True, label='r')
plt.arrow(0, 0, t_paralelo[0], t_paralelo[1], head_width=.05, color='r', linestyle='--', length_includes_head=True, label='t_para')
plt.arrow(t_paralelo[0], t_paralelo[1], t_perp[0], t_perp[1], head_width=.05, color='b', linestyle=':', length_includes_head=True, label='t_perp')
plt.axis('equal')
plt.legend()
plt.show()

print("\n--- Exercício 2-10: Debugging (Código com Bug proposital) ---")
# Bug: denominador t.t ao invés de r.r
t_paralelo_bug = r * (np.dot(t, r) / np.dot(t, t))
t_perp_bug = t - t_paralelo_bug

dot_check = np.dot(t_perp_bug, r)
print(f"Dot product da versão bugada: {dot_check:.4f}")
if not np.isclose(dot_check, 0):
    print(">>> O código falhou no teste de sanidade (não é ortogonal)!")
```

Exercícios 3-1 e 3-2 (Combinação Linear e Erros)

```python
print("--- Exercício 3-1: Combinação Linear com Listas ---")

lambdas = [1, 2, -3]
v1 = np.array([4, 5, 1])
v2 = np.array([-4, 0, -4])
v3 = np.array([1, 3, 2])
vetores = [v1, v2, v3]

w = np.zeros(vetores[0].shape)

for i in range(len(lambdas)):
    w = w + lambdas[i] * vetores[i]

print(f"Resultado via loop: {w}")

w_direto = lambdas[0]*vetores[0] + lambdas[1]*vetores[1] + lambdas[2]*vetores[2]
print(f"Resultado direto:   {w_direto}")
print(f"Iguais? {np.allclose(w, w_direto)}\n")


print("--- Exercício 3-2: Testando Erros de Dimensão ---")

# Caso 1: Vetor de dimensão diferente
vetores_erro = [v1, v2, v3, np.array([1, 2, 3, 4])] # Último vetor é R4
lambdas_erro = [1, 2, -3, 1]

try:
    w_erro = np.zeros(vetores_erro[0].shape)
    for i in range(len(lambdas_erro)):
        w_erro = w_erro + lambdas_erro[i] * vetores_erro[i]
except Exception as e:
    print(f"Erro detectado (Dimensões): {e}")

# Caso 2: Mais escalares que vetores
lambdas_extra = [1, 2, -3, 5]
try:
    w_erro = np.zeros(vetores[0].shape)
    for i in range(len(lambdas_extra)):
        w_erro = w_erro + lambdas_extra[i] * vetores[i]
except IndexError as e:
    print(f"Erro detectado (Índice): {e}")
```
